```python
import nltk
import string
nltk.download('punkt_tab')

def remove_punctuations_func(text):
    # Tokenize the text into individual words
    words = nltk.word_tokenize(text)
    # Remove punctuations from each word
    words_without_punctuations = [word for word in words if word not in string.punctuation]
    # Join the words back into a string
    text_without_punctuations = " ".join(words_without_punctuations)

    return text_without_punctuations

# Test the function
input_text = "Furthermore, I found the support really great. Paid about 180$ for it (5.7inch version, 4.8'' screen)."
result_text = remove_punctuations_func(input_text)
print("OUTPUT REMOVAL OF PUNCTUATIONS IN THE TEXT USING NLTK")
print("Original string:", input_text)
print("Modified string:", result_text)
```

```python
import re

def remove_punctuations_func(text):

    text_without_punctuations = re.sub(r'[^a-zA-Z0-9]', ' ', text)

    return text_without_punctuations

# Test the function
input_text = "Furthermore, I found the support really great. Paid about 180$ for it (5.7inch version, 4.8'' screen)."
result_text = remove_punctuations_func(input_text)
print("OUTPUT REMOVAL OF PUNCTUATIONS IN THE TEXT USING NLTK")
print("Original string:", input_text)
print("Modified string:", result_text)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string

# Example usage
input_string = "This is an example sentence to remove stop words from."
result_text = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result_text)
```

```python
import spacy
def remove_stop_words(string):
    # Load the spaCy English language model
    nlp = spacy.load('en_core_web_sm')

    # Tokenize the string into individual words
    doc = nlp(string)

    # Filter out stop words
    filtered_words = [token.text for token in doc if not token.is_stop]

    # Join the filtered words back into a string
    new_string = ' '.join(filtered_words)

    return new_string
```
